```python
import numpy as np
```

```python
first_array_from_list = np.array([1, 2, 3, 4])  # одномерный массив
second_array_from_list = np.array([[1, 2, 3], [4, 5, 6]])  # двумерный массив
```

```python
array_of_zeros = np.zeros((2, 3)) 
array_of_ones = np.ones(5) 
range_array = np.arange(0, 10, 0.5)  # массив от 0 до 10 с шагом 0.5
```

```python
# Демонстрация shape и dtype
print("Форма массива second_array_from_list:", second_array_from_list.shape)
print("Тип данных массива range_array:", range_array.dtype)
```

```python
element = second_array_from_list[1, 2]  # элемент второй строки и третьего столбца
subarray = second_array_from_list[:, 1:]  # подмассив всех строк, начиная со второго столбца
```

```python
print("Отдельный элемент:", element)
print("Подмассив:\n", subarray)
```
